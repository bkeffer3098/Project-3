```python
url = "https://raw.githubusercontent.com/albertyw/avenews/master/old/data/average-latitude-longitude-countries.csv"
import pandas as pd
df_latlng = pd.read_csv(url)
df_latlng
```

```python
data = "data.csv"
df = pd.read_csv(data)
df.head()
```

```python
column_names = df.columns.values.tolist()
column_names
```

```python
rename_df = df.rename(columns={"Location": "Country"})
rename_df
```

```python
merge_df = pd.merge(rename_df, df_latlng,  how = 'left', on="Country")
merge_df
```

```python
column_names = merge_df.columns.values.tolist()
column_names
```

```python
#merge_df.to_csv(r"C:\Users\Brian Keffer\GW Bootcamp\Project-3\merged_data.csv", index=False)
```

```python
final_data_df = merge_df.rename(columns={"Indicator": "HALE", "Period":"Year","Value": "Age Expectancy", "Dim1":"Sex"})
final_data_df
```

```python
# This needs to be done after renaming the columns appropriately for the columns we want. Turned into a DF
# After this export as a new csv for final use
final_data = final_data_df[["HALE", "Country", "Year", "Sex", "Age Expectancy", "Latitude", "Longitude"]]
final_data
```
