```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Set display options to show all rows and columns
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

# Specify the raw GitHub URL for the CSV file
github_url = 'https://raw.githubusercontent.com/bkeffer3098/Project-3/main/final_data.csv'

# Load the data from the CSV file
df = pd.read_csv(github_url)

# Randomly select the first 50 countries
random_countries = np.random.choice(df['Country'].unique(), size=50, replace=False)

# Filter the data for life expectancy at age 60 for men and women using loc
life_expectancy_men = df[(df['Sex'] == 'Male') & (df['Country'].isin(random_countries))]
life_expectancy_women = df[(df['Sex'] == 'Female') & (df['Country'].isin(random_countries))]

# Calculate mean life expectancy at age 60 for men and women by country
mean_life_expectancy_men = life_expectancy_men.groupby('Country')['Age Expectancy'].mean()
mean_life_expectancy_women = life_expectancy_women.groupby('Country')['Age Expectancy'].mean()

# Create a scatter plot
fig, ax = plt.subplots(figsize=(15, 8))
ax.scatter(
    mean_life_expectancy_men.index, mean_life_expectancy_men.values,
    label='Men', color='blue', alpha=0.7
)

ax.scatter(
    mean_life_expectancy_women.index, mean_life_expectancy_women.values,
    label='Women', color='orange', alpha=0.7
)

# Add error bars for standard deviation
ax.errorbar(
    mean_life_expectancy_men.index, mean_life_expectancy_men.values,
    yerr=life_expectancy_men.groupby('Country')['Age Expectancy'].std(),
    fmt='none', color='black', capsize=5
)

ax.errorbar(
    mean_life_expectancy_women.index, mean_life_expectancy_women.values,
    yerr=life_expectancy_women.groupby('Country')['Age Expectancy'].std(),
    fmt='none', color='black', capsize=5
)

ax.set_xlabel('Country')
ax.set_ylabel('Mean Life Expectancy at Age 60')
ax.set_title('Scatter Plot of Life Expectancy at Age 60 by Country for Men and Women (Random Selection of 50 Countries)')
ax.legend()
ax.tick_params(axis='x', rotation=90)  # Rotate x-axis labels for better visibility

plt.show()

```
