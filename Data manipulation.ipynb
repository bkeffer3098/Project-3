```python
url = "https://raw.githubusercontent.com/albertyw/avenews/master/old/data/average-latitude-longitude-countries.csv"
import pandas as pd
df_latlng = pd.read_csv(url)
df_latlng
```

```python
data = "data.csv"
df = pd.read_csv(data)
df.head()
```

```python
column_names = df.columns.values.tolist()
column_names
```

```python
rename_df = df.rename(columns={"SpatialDimValueCode": "Country Code"})
df_latlng = df_latlng.rename(columns={"ISO 3166 Country Code": "Country Code"})
df_latlng
```

```python
merge_df = pd.merge(rename_df, df_latlng,  how = 'left', on="Country Code")
merge_df
```

```python
column_names = merge_df.columns.values.tolist()
column_names
```
